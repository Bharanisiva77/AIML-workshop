```python
import chromadb
```

```python
from sentence_transformers import SentenceTransformer
```

```python
print("Loading embedding model...")
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
print("Model loaded successfully")
```

```python
client = chromadb.Client()
```

```python
collection = client.get_or_create_collection(
    name="food_collection"
)
```

```python
print("Collection created successfully")
```

```python
food_items = [

    "Pizza is an Italian food made with cheese and tomato.",

    "Burger is a fast food made with bun and patty.",

    "Biryani is a spicy rice dish popular in India.",

    "Dosa is a South Indian breakfast food.",

    "Idli is a healthy steamed South Indian food.",

    "Pasta is an Italian dish made using noodles.",

    "Fried rice is made with rice and vegetables.",

    "Ice cream is a frozen sweet dessert.",

    "Pani puri is a famous Indian street food.",

    "Sandwich is made with bread and vegetables.",

    "Chocolate cake is a sweet bakery dessert.",

    "Noodles are popular Chinese food.",

    "Samosa is a fried Indian snack.",

    "Parotta is a layered flatbread from South India.",

    "Chicken curry is a spicy non vegetarian dish.",

    "Salad is a healthy food made with vegetables.",

    "French fries are deep fried potato snacks.",

    "Tacos are traditional Mexican food.",

    "Momos are steamed dumplings.",

    "Paneer butter masala is a North Indian curry."

]
```

```python
print("Food data loaded successfully")
```

```python
print("Generating embeddintgs...")
```

```python
embeddings = model.encode(food_items)
```

```python
print("Embeddings generated successfully")
```

```python
ids = [str(i) for i in range(len(food_items))]
```

```python
collection.add(
    documents=food_items,
    embeddings=embeddings.tolist(),
    ids=ids
)
```

```python
print("Food data stored in ChromaDB successfully")
```

```python
def search_food_data(user_query, results_count=3):

    print("\n")
    print("===================================")
    print("USER QUERY :", user_query)
    print("===================================")

    # Convert query into embedding
    query_embedding = model.encode([user_query]).tolist()

    # Search similar food items
    results = collection.query(
        query_embeddings=query_embedding,
        n_results=results_count
    )

    print("\nTOP MATCHING FOOD RESULTS:\n")

    for i, document in enumerate(results['documents'][0]):

        print(f"{i+1}. {document}")

    print("\n")
```

```python
search_food_data("healthy food")

search_food_data("Indian spicy food")

search_food_data("dessert item")

search_food_data("South Indian breakfast")

search_food_data("fried snacks")
```

```python
while True:




    user_input = input("Ask Food Question (or type exit): ")

    if user_input.lower() == "exit":
        print("Program Ended")
        break

    search_food_data(user_input)
```
